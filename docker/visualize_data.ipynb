### Libraries imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Data Load

```python
csv_path = '/home/mikemadeira_aux/Documents/Projects/Intelligent-Debt-Collection/docker/bank_debt_data/' 

df = pd.read_csv(csv_path+'client_data.csv')
print(df.head())
```

### Data validation

```python
# Categorical data visualization
data = df.annual_income

sns.displot(data)
plt.title('Histogram of Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Categorical data visualization
data = df.number_credits_loans

num_bins = len(data.unique())

sns.displot(data,bins=num_bins)
plt.title('Histogram of Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
df_scatter = df[df['annual_income']<80000]

sns.scatterplot(data=df_scatter,x=df_scatter.annual_income,y=df_scatter.number_credits_loans)
```
